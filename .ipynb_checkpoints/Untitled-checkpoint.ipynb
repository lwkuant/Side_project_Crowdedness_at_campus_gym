```python
### The relationships between each categorical variable
def cal_chi_table(df, col_name=[]):
    from scipy.stats import chi2_contingency
    
    chi_mat = np.zeros([len(col_name), len(col_name)])
    
    for i in range(len(col_name)):
        for j in range(len(col_name)):
            if i == j:
                chi_mat[i, j] = 0
            else:
                chi_mat[i, j] = 1 - chi2_contingency(pd.crosstab(df[col_name[i]], df[col_name[j]]))[1]
    return chi_mat

col_name = ['day_of_week', 'is_weekend', 'is_holiday', 'is_start_of_semester', 'is_during_semester',
                              'month', 'hour']
sns.heatmap(cal_chi_table(df, col_name),  xticklabels=col_name, yticklabels=col_name, square=True, linewidth=1)    
plt.title('Relationships within Categorical Features', y=1.05)
```

Many categorical features are highly-related.

## Modeling 

```python
df.info()
```

```python
### Split the dataset to training set and test set
seed = 100
np.random.seed(seed)

from sklearn.model_selection import train_test_split
df_tr, df_test = train_test_split(df, test_size=0.2, random_state=seed)

df_tr.index = range(len(df_tr))
df_test.index = range(len(df_test))

print(df_tr.shape)
```

```python
### Get ready for modeling

cat_col = ['day_of_week', 'is_weekend', 'is_holiday', 'is_start_of_semester', 'is_during_semester', 'month', 'hour']
num_col = ['temperature']

def model_process(df, df_tr, num_col=num_col, cat_col=cat_col):
    
    ### Coonstruct the one hot encoding 
    def label_encoder(df, col):
        from sklearn import preprocessing
        le = preprocessing.LabelEncoder()
        le.fit(df[col].values)

        return le

    def one_hot(df, col, label_encoder):
        from sklearn import preprocessing
        vec = label_encoder.transform(df[col].values).reshape([len(df[col]), 1])
        ohc = preprocessing.OneHotEncoder(sparse=False, n_values=len(label_encoder.classes_))
        ohc_df = ohc.fit_transform(vec)

        ohc_df = pd.DataFrame(ohc_df, columns=[col+'_'+str(cat) for cat in label_encoder.classes_], index=range(len(df[col])))
        return ohc_df
    
    X = df.ix[:, num_col]
    
    for cat in cat_col:
        X = pd.concat([X, one_hot(df, cat, label_encoder(df_tr, cat))], axis=1)
        
    y = df['number_people']
    
    return X, y
```

```python
X, y = model_process(df_tr, df_tr)

print(X.shape)
```

```python
X = X.values
y = y.values
```

#### Find good features

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X, y)
```

```python
df_feature, _ = model_process(df_tr, df_tr)
print(np.array(df_feature.columns)[np.argsort(model.feature_importances_)][::-1])
```

#### Linear Regression (baseline)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kf = KFold(5, random_state=seed, shuffle=True)

for fold in kf.split(X):
    model = LinearRegression()
    model.fit(X[fold[0]], y[fold[0]])
    print(np.sqrt(mean_squared_error(y[fold[1]], model.predict(X[fold[1]]))))
```

#### Ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kf = KFold(5, random_state=seed, shuffle=True)

for fold in kf.split(X):
    model = Ridge()
    model.fit(X[fold[0]], y[fold[0]])
    print(np.sqrt(mean_squared_error(y[fold[1]], model.predict(X[fold[1]]))))
```

#### Elastic net

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kf = KFold(5, random_state=seed, shuffle=True)

for fold in kf.split(X):
    model = ElasticNet()
    model.fit(X[fold[0]], y[fold[0]])
    print(np.sqrt(mean_squared_error(y[fold[1]], model.predict(X[fold[1]]))))
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kf = KFold(5, random_state=seed, shuffle=True)

for fold in kf.split(X):
    model = RandomForestRegressor(max_features=8, n_estimators=500)
    model.fit(X[fold[0]], y[fold[0]])
    print(np.sqrt(mean_squared_error(y[fold[1]], model.predict(X[fold[1]]))))
```

training each model would take about 30s for setting: (max_features=8, n_estimators=500)

#### xgbregressor

```python
from xgboost import XGBRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

kf = KFold(5, random_state=seed, shuffle=True)

for fold in kf.split(X):
    model = XGBRegressor()
    model.fit(X[fold[0]], y[fold[0]])
    print(np.sqrt(mean_squared_error(y[fold[1]], model.predict(X[fold[1]]))))
```

## Grid Search to find the best set of parameters

```python
### Grid search to choose the best model
import time
start = time.time()

from sklearn.ensemble import RandomForestRegressor
from sklearn.grid_search import GridSearchCV
param_grid = {'n_estimators': [200, 300, 500], 'max_features': ['sqrt', 'log2'], 'max_depth': [5, 10, 15]}
model = RandomForestRegressor(n_jobs=-1)
grid = GridSearchCV(model, param_grid, cv=5, n_jobs=-1)
grid.fit(X, y)

total = time.time() - start
print('Execution time: %s' %(str(total)))
print('done')
```
